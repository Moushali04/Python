```python
import pandas as pd     # data processing, CSV file I/O (e.g. pd.read_csv)
import numpy as np      # linear algebra
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("/content/prices.csv")
df
```

```python
!ls
```

```python
%matplotlib inline
plt.xlabel('area(sqr ft)')
plt.ylabel('price(US$)')
plt.scatter(df.area, df.price, color='blue', marker='.')
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
area = 3200
area_reshaped = np.array(area).reshape(1, -1)  # Reshape scalar to 2D array
predict = reg.predict(area_reshaped)
print('Price for {} area is : {}'.format(area, predict))
```

```python
coef = reg.coef_
intercept = reg.intercept_
print('Coefisient : {}'.format(coef))
print('Intercept : {}'.format(intercept))
```

```python

coef * area + intercept
```

```python

plt.xlabel('area', fontsize=20)
plt.ylabel('price', fontsize=20)
plt.scatter(df.area, df.price, color='red', marker='+')
plt.plot(df.area, reg.predict(df[['area']]), color='blue')
```
